```python
import os
import pandas 
import pickle as pkl 
import numpy as np
import matplotlib.pyplot as plt

TRAJ_DIR = "idil_gym/envs/test_gen"
```

# `MultiGoals2D_3`

```python
with open(os.path.join(TRAJ_DIR, "MultiGoals2D_3-v0_2.pkl"), "rb") as f:
    data = pkl.load(f)
```

```python
fig ,ax  = plt.subplots(1, 1, figsize=(7,7))

for idx in range(len(data["states"])):

    sample_traj = np.array(data["states"][idx])
    # plot the trajectory
    ax.plot(sample_traj[:,0], sample_traj[:,1])
```
